```python
#Write a python program to demonstrate the application of Histogram equalization for a image and
#Histogram matching for source and reference images
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.exposure import match_histograms

# Read grayscale images
src = cv2.imread("image.jpg", 0)        # Source image
ref = cv2.imread("image.jpg", 0)    # Reference image

# Histogram Equalization (inbuilt OpenCV)
hist_eq = cv2.equalizeHist(src)

# Histogram Matching (inbuilt skimage)
hist_match = match_histograms(src, ref,channel_axis=None)

# ---------------- IMAGE DISPLAY (Side by Side) ----------------

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(src, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(hist_eq, cmap='gray')
plt.title("Histogram Equalized")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(hist_match, cmap='gray')
plt.title("Histogram Matched")
plt.axis("off")

plt.show()

# ---------------- HISTOGRAM DISPLAY (Side by Side) ----------------

plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.hist(src.ravel(), 256, [0,256])
plt.title("Original Histogram")

plt.subplot(1, 3, 2)
plt.hist(hist_eq.ravel(), 256, [0,256])
plt.title("Equalized Histogram")

plt.subplot(1, 3, 3)
plt.hist(hist_match.ravel(), 256, [0,256])
plt.title("Matched Histogram")

plt.show()
```
